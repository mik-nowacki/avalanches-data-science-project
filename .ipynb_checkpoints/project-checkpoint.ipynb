**Description**

([source](https://opendata.swiss/en/dataset/fatal-avalanche-accidents-in-switzerland-since-1995-1996))
<br>Attention: this data is not updated after 2022 anymore. This data collection contains information concerning all accidents by snow avalanches causing at least one fatality in Switzerland. The data set commences on 01/10/1995. After the completion of a hydrological year, the new data is added. The following information is provided:

- **avalanche.id** = avalanche identifier  
- **date** = date of the accident (YYYY-MM-DD)
- **date.quality** = accuracy of the date in range of days before and after
- **hydrological.year** = 12-month period from 1 October to 30 September
- **canton** = Cantons of Switzerland (Member states of the Swiss Confederation)
- **local.name** = name of the locality  
- **start.zone.coordinates.x/y** = start zone of the avalanche (Swiss coordinate system, approximately in the middle of start zone)  
- **coordinates.quality** = accuracy of the coordinates in meters  
- **start.zone.elevation** = elevation (in meters above sea level, approximately in the middle of start zone)  
- **start.zone.slope.aspect** = slope aspect (main orientation of start zone)  
- **start.zone.inclination** = slope inclination (in degree, steepest point within start zone)  
- **forecasted.dangerlevel** = forecasted avalanche danger level (1-5)  
- **number.dead** = number of dead persons  
- **number.caught** = number of caught persons  
- **number.fully.buried** = number of fully buried persons  
- **activity** = activity/location of the accident party at the time of the incident

### 2. Number of accidents per month

```python
from datetime import datetime
avalanches['date'] = pd.to_datetime(avalanches['date'], format="%Y-%m-%d")
avalanches['month'] = avalanches['date'].dt.month
accidents_by_month = avalanches.groupby(by='month', as_index=False)[ACCIDENT_TYPES].sum()

hydrological_year = list(range(9, 13)) + list(range(1, 9))
accidents_by_month = accidents_by_month.set_index('month').reindex(hydrological_year).reset_index()


x = accidents_by_month['month']
X_axis = np.arange(len(accidents_by_month))

fig2, ax2 = plt.subplots(figsize=(8,5))

ax2.bar(X_axis - 0.25, accidents_by_month['number.caught'], 0.25, label='Caught')
ax2.bar(X_axis, accidents_by_month['number.fully.buried'], 0.25, label='Fully Buried')
ax2.bar(X_axis + 0.25, accidents_by_month['number.dead'], 0.25, label='Dead')
ax2.set_xticks(range(len(x)), x)
ax2.set_xlabel('Month')
ax2.set_ylabel('Number of people')
ax2.set_title('2. Accidents per month')
ax2.legend(loc='upper right')

fig2.show()
```

Plot 2 shows the number of accidents per month. The dataframe has been rearranged so that the months are ordered according to the hydrological year (from September to August). By doing so, it is easier to observe that the data follows a **Gaussian ditribution**, which is naturally true. February is usually the peak season, where the snow has been built up and the temperature has already reached its lowest point. It is also popular to go skiing in that period due to winter holidays in many countries.

### 3. Number of accidents per activity

```python
accidents_by_activity = avalanches.groupby(['activity'], as_index=False)[ACCIDENT_TYPES].sum()
accidents_by_activity['number.combined'] = accidents_by_activity[ACCIDENT_TYPES].sum(1)

# Combine "building" into "other, mixed or unknown"
# Step 1: Identify rows
building_row = accidents_by_activity[accidents_by_activity['activity'] == 'building']
other_unknown_row = accidents_by_activity[accidents_by_activity['activity'] == 'other, mixed or unknown']

# Step 2: Sum the relevant columns
combined_row = other_unknown_row.copy()
for col in ['number.dead', 'number.caught', 'number.fully.buried', 'number.combined']:
    combined_row[col] = building_row[col].values[0] + other_unknown_row[col].values[0]

# Step 3: Update the "other, mixed or unknown" row
accidents_by_activity.loc[
    accidents_by_activity['activity'] == 'other, mixed or unknown', 
    ['number.dead', 'number.caught', 'number.fully.buried', 'number.combined']
] = combined_row[['number.dead', 'number.caught', 'number.fully.buried', 'number.combined']].values

accidents_by_activity = accidents_by_activity[accidents_by_activity['activity'] != 'building'].reset_index(drop=True)

fig2, ax2 = plt.subplots()
ax2.pie(
    accidents_by_activity['number.combined'].astype(int),
    labels = accidents_by_activity['activity'], 
    autopct='%1.1f%%',  # add percentage labels
    startangle=90
)
plt.title('3. Accidents by Activity')
plt.show()
```

According to plot 3 the most accidents happen during ski-touring. This type of skiing involves a special kind of ski and boots which allow both for skiing and walking up hill in difficult terrain outside skiing resorts. Sometimes the walk-up takes several hours and involves entering danger zones which explains why touring accidents are the most frequent. 
<br> Offpist is the second dominant activity which involves using a lift to go uphill and then ski down the mountain outside a designated area. Less exposure time on the unprepared areas makes the accidents less frequent.  

### 4. Correlation between start zone elevation and the number of accidents

```python
from sklearn import linear_model
import math

elevation_data = avalanches[['start.zone.elevation'] + ACCIDENT_TYPES].groupby(['start.zone.elevation'], as_index=False).sum()
elevation_data['number.combined'] = elevation_data[ACCIDENT_TYPES].sum(axis=1)

# Move data into clusters that group expeditions according to their starting elevation.
height = 750  # starting elevation (all expeditions starting from less than 750m)
step = 250
elevation_height = [height]
elevation_accidents = [0]

for i in range(0, len(elevation_data)):
    current_elevation = elevation_data.iloc[i]['start.zone.elevation']
    if  current_elevation <= height:
        elevation_accidents[-1] += elevation_data.iloc[i]['number.combined']
    else:
        height = math.ceil(current_elevation / step) * step
        elevation_height.append(height)
        elevation_accidents.append(elevation_data.iloc[i]['number.combined'])

    
elevation_grouped = pd.DataFrame(data = {'start.zone.elevation': elevation_height, 'number.combined': elevation_accidents})

regr = linear_model.LinearRegression()
regr.fit(np.array(elevation_grouped['start.zone.elevation']).reshape(-1,1), np.array(elevation_grouped['number.combined']).reshape(-1,1))
y_reg = regr.predict(np.array(elevation_grouped['start.zone.elevation']).reshape(-1,1))

plt.scatter(elevation_grouped['start.zone.elevation'], elevation_grouped['number.combined'], label='Data')
plt.plot(elevation_grouped['start.zone.elevation'], y_reg, 'r', label='Linear Regression')
plt.xlabel('Start Zone Elevation')
plt.ylabel('Number of Combined Accidents')
plt.title('4. Accidents vs. Elevation')
plt.legend()
plt.grid(True)
```

Plot 4 shows correlation between the elevation and the number of accidents. When thinking about it, it is obvious that the higher one starts the expedition the more dangerous it gets. It is generally true, and we can observe an almost exponential growth in accidents count up until 2500m. After that point correlation suddenly drops. This may be explained by the fact that there is signifficantly less people who are experienced enough to venture into the higher parts of the Alps in the winter. There are also fewer spots to start an expedition from the higher one goes. 

### 5. Correlation between the slope aspect and the number of accidents

```python
accidents_by_slope_aspect = avalanches[['start.zone.slope.aspect'] + ACCIDENT_TYPES].copy()
# Drop rows that the slope aspect is NaN
accidents_by_slope_aspect.dropna(subset=['start.zone.slope.aspect'], inplace=True)
# Reduce the number of directions of slopes to 8 aspects (instead of 16) to improve radability.
# (most of the maps show only 8 aspects)
accidents_by_slope_aspect['start.zone.slope.aspect'] = [row[1:] if len(row) > 2 else row for row in accidents_by_slope_aspect['start.zone.slope.aspect']]
# Count accidents by the slope aspect
accidents_by_slope_aspect = accidents_by_slope_aspect.groupby(['start.zone.slope.aspect'], as_index=False).sum()
accidents_by_slope_aspect['number.combined'] = accidents_by_slope_aspect[ACCIDENT_TYPES].sum(axis=1)


plt.bar(accidents_by_slope_aspect['start.zone.slope.aspect'], accidents_by_slope_aspect['number.combined'])
plt.xlabel('Slope aspect')
plt.ylabel('Number of accidents')
plt.title('5. Number of accidents by the slope aspect')
```

It is generally known that the north-facing slopes tend to have greater avalanche danger, due to deeper snow and increased layering from wind and a lack of sunlight ([source](https://www.skimag.com/adventure/is-your-slope-safe/)). Plot 5 confirms it with North-West and North-East having the most accidents.

### Hypothesis 1. 
At the 95% confidence level, is there enough data to support the idea that there is a significant difference in the number of accidents between periods 1995-2010 and 2011 - 2021 during peak season (February)?<br>

**Null Hypothesis ($H_0$)**: There is no significant difference in the number of accidents between the two periods (1995–2010 and 2011–2021) during peak season.<br>
- $H_0$: $μ_1 = μ_2$

**Alternative Hypothesis ($H_A$)**: There is a significant difference in the number of accidents betwen the two periods.<br>
- $H_A$: $μ_1$≠$μ_2$

These two samples are independent and with different variances. To perform the hypothesis test the two-sample t-test will be used.

```python
from scipy.stats import t, ttest_ind

alpha = 0.05
avalanches['year'] = avalanches['date'].dt.year
period1 = avalanches[(avalanches['year']<=2010) & (avalanches['month']==2)].copy()
period2 = avalanches[(avalanches['year']>2010) & (avalanches['month']==2)].copy()

period1['number.combined'] = period1[ACCIDENT_TYPES].sum(axis=1)
period2['number.combined'] = period2[ACCIDENT_TYPES].sum(axis=1)

n1 = len(period1)
n2 = len(period2)

mean1 = period1['number.combined'].mean()
mean2 = period2['number.combined'].mean()

std1 = period1['number.combined'].std()
std2 = period2['number.combined'].std()

# t_test =  (mean1 - mean2)/ np.sqrt(std1**2/n1 + std2**2/n2)
t_test = ttest_ind(period1['number.combined'], period2['number.combined'], equal_var=False)
# dof = (std1**2/n1 + std2**2/n2)**2 / (((std1**2/n1)**2/(n1-1)) + ((std2**2/n2)**2/(n2-1)))

# null_dist = t.pdf(t_test,dof)
# p_value = 2 * min(t.cdf(t_test, df = dof), 1 - t.cdf(t_test, df = dof))
print('p-value =', t_test.pvalue)
```

The p-value of 0.083 is greater than alpha = 0.05 which means we fail to reject the null hypothesis. There is no strong evidence to suggest a significant difference in the number of avalanche accidents during February between the two periods (1995–2010 and 2011–2021).
